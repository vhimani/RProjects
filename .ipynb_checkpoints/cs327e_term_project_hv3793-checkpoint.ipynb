# Define a Project

## Load, describe, and clean the dataset

The boston dataset contains various geographic and demographic variables to predict housing prices.

```python
# Load the boston dataset
X, y = load_boston(return_X_y=True) 
boston_df_X = pd.DataFrame(X, columns=load_boston().feature_names)
boston_df_y = pd.DataFrame(y)
boston_df = boston_df_X.merge(boston_df_y, left_index=True, right_index=True)
boston_df = boston_df.rename(columns={0:'MEDV'})
boston_df.head()
```

```python
boston_df.shape
```

The 14 variables are defined as follows:
1. CRIM: per capita crime rate by town 
2. ZN: proportion of residential land zoned for lots over 25,000 sq.ft.
3. INDUS: proportion of non-retail business acres per town
4. CHAS: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
5. NOX: nitric oxides concentration (parts per 10 million)
6. RM: average number of rooms per dwelling
7. AGE: proportion of owner-occupied units built prior to 1940
8. DIS: weighted distances to five Boston employment centres
9. RAD: index of accessibility to radial highways
10. TAX: full-value property-tax rate per 10,000 dollars
11. PTRATIO: pupil-teacher ratio by town
12. B: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
13. LSTAT: % lower status of the population
14. MEDV: Median value of owner-occupied homes in 1000 dollars

The total observations for my dataset is less than 2000 and there are currently no necessary cleaning operations to perform.

# Implementation

### Pre-Regression Analysis
In this part of the project, I will check for correlations and linearity of the variables. I will construct a correlation matrix to check for correlations and scatterplots to evaluate if the variables look reasonably straight for regression analysis. 

```python
# Checking correlations
corr = boston_df.corr()
f, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
```

We see that these 4 combinations of variables are highly positively correlated (corr > 0.7):
- TAX and RAD
- NOX and INDUS
- NOX and AGE
- MEDV and RM

We see that these 4 combinations of variables are highly negatively correlated (corr < -0.7):
- MEDV and LSTAT
- DIS and AGE
- DIS and NOX
- DIS and INDUS

Since I will further check the variance inflation factor (VIF) of these variables when I run the multiple linear regression, I will then drop variables accordingly.

```python
# Plot correlation plots to assess linearity of the data
x_vars=list(boston_df_X.columns)
g = sns.FacetGrid(pd.DataFrame(x_vars), col=0, col_wrap=5, sharex=False)
for ax, x_var in zip(g.axes, x_vars):
    sns.scatterplot(data=boston_df, x=x_var, y=boston_df['MEDV'], ax=ax)
g.tight_layout()
```

Since the CHAS variable is a dummy variable, the scatterplot has points concentrated at either 0 or 1. The RAD variable has discrete values either less than 10 or close to 25, which explains the scatter of the points in its plot. We see that most of the plots look reasonably straight with varying degree of correlation (low vs moderate vs high), so I will proceed further with my regression analysis. 

### Linear Regression: Evaluating Model Construction

I use two libraries to implement linear regression. In the first part, I use statsmodels because I am interested in learning about the significance of the variables, and through the summary method I can easily access that.

```python
# Implementing ordinary least squares regression using statsmodels library
ols = sm.OLS(boston_df_y, sm.add_constant(boston_df_X)).fit()
print(ols.summary())
```

The R-squared of the data is 0.741. This means that the model explains 74.1% of the variability in the data. 
<br/>The Adjusted R-squared of the model is 0.734. We will use the adjusted R-squared for comparsion further as we drop variables as R-squared is strictly higher for more variables.

```python
# Calculate MAE and RMSE
mae = np.absolute(np.subtract(ols.model.endog, ols.fittedvalues)).mean()
rmse = np.sqrt(((ols.model.endog - ols.fittedvalues)**2).mean())
print('The MAE of the model is', mae, 'and the RMSE is', rmse)
```

The MAE of the model is 3.2709, and the RMSE is 4.6792.

Additionally, the variables INDUS and AGE are highly insignificant (p > 0.05). I will first drop these variables from my model.

```python
# Dropping insignificant variables
boston_df_X1 = boston_df_X.drop(columns=['INDUS','AGE'])
```

```python
# Implementing linear regression with only significant variables
ols = sm.OLS(boston_df_y, sm.add_constant(boston_df_X1)).fit()
print(ols.summary())
```

Dropping the insignificant variables slightly increases the adjusted R-squared of the model. The adjusted R-squared of the reduced model is 73.5%.

```python
# Calculate MAE and RMSE
mae = np.absolute(np.subtract(ols.model.endog, ols.fittedvalues)).mean()
rmse = np.sqrt(((ols.model.endog - ols.fittedvalues)**2).mean())
print('The MAE of the model is', mae, 'and the RMSE is', rmse)
```

The MAE of the model is 3.2720 and the RMSE is 4.6797.
<br/>In this case, MAE increases slighly by 0.0012. RMSE also increases by 0.0005.

We still need to check for multicollinearity. I will use the variance inflation factor (VIF) to look for highly correlated variables. 

```python
df_w_constant = sm.add_constant(boston_df_X)
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(df_w_constant.values, i) for i in range(df_w_constant.values.shape[1])]
vif["features"] = df_w_constant.columns
vif
```

```python
# Dropping all variables with VIF over 10 as it's an indication of high collinearity
vif = vif[vif['VIF'] <= 10]
vif.reset_index(drop=True)
```

```python
# Dropping columns with high VIF
boston_df_X1 = boston_df_X[list(vif['features'].unique())]
boston_df_X1.head()
```

Since all the variables are preserved after the multicollinearity check, we will get the same results as the first model for all of our regression metrics.

**Conclusion**: Although overall the adjusted-R squared is slightly higher for the second model (model with only significant variables), the MAE and RMSE are larger. This is an example of the interpretability vs predictability tradeoff. Since I want to reduce predictibilty errors in my further analysis, I will stick with my first complete model with all the variables which has less error rates for next analysis.

### Linear Regression: Testing Predictibility

For this part of the analysis, I will split my data into training and test sets. I will train my model on the training sets and compare the regression metrics by fitting it on the test set. Since I am more familiar with sklearn through class for splitting and training models, I will use it in this section of my analysis.

```python
# Split the dataset into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(boston_df_X, boston_df_y, test_size = 0.33, random_state = 1)
```

```python
# Check the shape of training and test sets 
print(X_train.shape)
print(X_test.shape)
print(boston_df_X.shape)
print(y_train.shape)
print(y_test.shape)
print(boston_df_y.shape)
```

```python
# Train the model
model = LinearRegression()
model.fit(X_train, y_train)
print(model.intercept_)
print(model.coef_)
```

```python
# Regression metrics for the train set
r_squared = model.score(X_train, y_train)
rmse = (np.sqrt(mean_squared_error(y_train, model.predict(X_train))))
mae = np.absolute(np.subtract(y_train, model.predict(X_train))).mean()
print('R-squared:', r_squared, 'RMSE:', rmse, 'MAE:', mae)
```

The R-squared of the training model is 0.722, or it explains 72.2% variability in the model. 
<br/>The RMSE is 4.79 and MAE is 3.28. 
<br/>The R-squared is smaller and the errors are larger than the complete model, which maskes sense as we are using lesser observations for the training set.

```python
# Regression metrics for the test set
r_squared = model.score(X_test, y_test)
rmse = (np.sqrt(mean_squared_error(y_test, model.predict(X_test))))
mae = np.absolute(np.subtract(y_test, model.predict(X_test))).mean()
print('R-squared:', r_squared, 'RMSE:', rmse, 'MAE:', mae)
```

The R-squared of the testing model is 0.765, or it explains 76.5% variability in the model. 
<br/>The RMSE is 4.54 and MAE is 3.41. 
<br/>The R-squared is larger than the testing model, which suggests that our testing model overall does well on predictions. MAE is smaller the RMSE is larger in this model, which is surprising as they seemed to have been moving in sync in the previous models. The difference only explains that there is some variation in the magnitude of the errors.

```python
# Check residual plot
plt.scatter(model.predict(X_test), (y_test - model.predict(X_test)))
plt.show()
```

The residuals show a horizontal direction, a shapeless form, and roughly equal scatter for all predicted values. The linear model appears to be appropriate.
